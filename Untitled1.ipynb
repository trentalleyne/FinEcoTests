```python
# Prices
prices = pd.read_csv('data/merged house prices.csv') #thanks to Peter Abelson, ABS, Domain
prices = prices.set_index(prices['Year'])
prices = prices.drop('Year', axis = 1)
```

```python
prices = prices[5:]
```

```python
# Incomes
MIM = pd.read_csv('data/MedianIncomes.csv')
MIM = MIM.set_index(MIM['Date']).astype(float)
MIM = MIM.drop('Date', axis = 1)
MIF = pd.read_csv('data/MedianIncomesFemales.csv')
MIF = MIF.set_index(MIF['Date']).astype(float)
MIF = MIF.drop('Date', axis = 1)
```

```python
#dealing with missing values via interpolation
for col in MIM.columns:
    MIM[col] = MIM[col].interpolate()
    MIF[col] = MIF[col].interpolate()
```

```python
MIH = MIM + MIF
```

```python
MIH = MIH*4 # working in months
```

```python
excel_file = 'data/constInputs.xlsx'
sheet_names = [0, 1, 2] 
constInputs = {}
for sheet in sheet_names:
    constInputs[sheet] = pd.read_excel(excel_file, sheet_name=sheet)
    constInputs[sheet] = constInputs[sheet].set_index(constInputs[sheet]['Year'])
    constInputs[sheet].drop('Year', axis = 1)
```

```python
excel_file = 'data/variedyears.xlsx'
sheet_names = [0, 1, 2] 
varyInputs = {}
for sheet in sheet_names:
    varyInputs[sheet] = pd.read_excel(excel_file, sheet_name=sheet)
    varyInputs[sheet] = varyInputs[sheet].set_index(varyInputs[sheet]['Year'])
    varyInputs[sheet].drop('Year', axis = 1)
```

```python
excel_file = 'data/varyInputsYearlyExp.xlsx'
sheet_names = [0, 1, 2] 
varyInputsYearlyExp = {}
for sheet in sheet_names:
    varyInputsYearlyExp[sheet] = pd.read_excel(excel_file, sheet_name=sheet)
    varyInputsYearlyExp[sheet] = varyInputsYearlyExp[sheet].set_index(varyInputsYearlyExp[sheet]['Year'])
    varyInputsYearlyExp[sheet].drop('Year', axis = 1)
```
