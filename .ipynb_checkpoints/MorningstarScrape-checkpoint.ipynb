```python
import pandas as pd
import numpy as np
from numpy import nan

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

import re
```

```python
urlForm = 'http://financials.morningstar.com/{}/{}.html?t={}&region=usa&culture=en-US'
dfForm = 'df_{}'
symbols = ["amzn", "aapl", "fb", "ibm", "msft"]
#symbols = ["amzn"]
```

```python
df = {}
for ticker in symbols:
    df[ticker] = pd.DataFrame()
```

```python
for i, symbol in enumerate(symbols):
    df[symbol] = pd.DataFrame({'year ending': ['info', 'net income', 'interest expense', 'total assets', 'short term debt', 'long term debt', 'FFCF']})
    df[symbol] = df[symbol].set_index('year ending')
    
    s = Service(r'C:\Users\trent\.wdm\drivers\chromedriver\win32\96.0.4664.45\chromedriver.exe')
    driver = webdriver.Chrome(service=s)
    
    elements = []
    elements.append(symbol) #0
    
    #Income statement scrape
    url = urlForm.format("income-statement", "is", symbol)
    
    driver.get(url)
    time.sleep(2)

    years = "//div/div[@id='Year']"
    info = "//div/div[@id='unitsAndFiscalYear']"
    netIncome = "//div/div[@id='data_i80']"
    interestExpense = "//div/div[@id='data_i51']"
    
    elements.append(re.findall('(.*)\n', (driver.find_element(By.XPATH, years).text))) #1
    elements.append(re.findall('(.*)', (driver.find_element(By.XPATH, info).text))) #2
    elements.append(re.findall('(.*)\n', (driver.find_element(By.XPATH, netIncome).text))) #3
    elements.append(re.findall('(.*)\n', (driver.find_element(By.XPATH, interestExpense).text))) #4
    
    #table headers
    for i in elements[1]:
        if i != '':
            df[symbol][i] = np.nan
            
    df[symbol].iloc[0, [0]] = elements[2][2]
    
    #Balance sheet scrape
    url = urlForm.format('balance-sheet', "bs", symbol)

    driver.get(url)
    time.sleep(2)
    
    totalAssets = "//div/div[@id='data_tts1']"
    shortDebt = "//div/div[@id='data_i41']"
    LongDebt = "//div/div[@id='data_i50']"
    
    elements.append(re.findall('(.*)', (driver.find_element(By.XPATH, totalAssets).text))) #5
    elements[5] = list(filter(('').__ne__, elements[5]))
    elements.append(re.findall('(.*)', (driver.find_element(By.XPATH, shortDebt).text))) #6
    elements[6] = list(filter(('').__ne__, elements[6]))
    elements.append(re.findall('(.*)', (driver.find_element(By.XPATH, LongDebt).text))) #7
    elements[7] = list(filter(('').__ne__, elements[7]))
    
    #Cash flow statement scrape
    url = urlForm.format('cash-flow', "cf", symbol)

    driver.get(url)
    time.sleep(2)
    
    FFCF = "//div/div[@id='data_i97']"
    elements.append(re.findall('(.*)\n', (driver.find_element(By.XPATH, FFCF).text))) #8
    
    #data to df
    row = 0
    for i in df[symbol].index:
        if row > 0:
            try:
                df[symbol].iloc[row] = elements[2+row]
            except:
                pass
        row += 1

driver.quit()
```
