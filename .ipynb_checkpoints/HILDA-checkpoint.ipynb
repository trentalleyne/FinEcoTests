### Education and Earnings

```python
# create a dataset with waveid, id, earnings, education
clean = hilda[['wave', 'hhrhid', 'rpevown', 'opnum', 'wscei', 'tifditp', 'tifditn', 'edhigh1']]
```

```python
clean.loc[clean['opnum'] == '[-1] Not asked', 'opnum'] = np.nan
clean.loc[clean['opnum'] == '[-4] Refused/Not stated'] = np.nan
clean.loc[clean['opnum'] == '[-3] Dont know', 'opnum'] = np.nan
clean['opnum'] = clean['opnum'].apply(pd.to_numeric, errors='coerce')
```

```python
education_mapping = {
    '[9] Year 11 and below': 1,
    '[8] Year 12': 2,
    '[5] Cert III or IV': 3,
    '[4] Adv diploma, diploma': 4,
    '[3] Bachelor or honours': 5,
    '[2] Grad diploma, grad certificate': 6,
    '[1] Postgrad - masters or doctorate': 7
}
clean['education'] = clean['edhigh1'].map(education_mapping)
```

```python
owns_prop_mapping = {
    '[1] Yes': 1,
    '[2] No': 0,
    '[-4] Refused/Not stated': np.nan,
    '[-3] Dont know': np.nan,
}
clean['owns property'] = clean['rpevown'].map(owns_prop_mapping)
```

```python
clean.rename(columns={'wscei': 'weekly work income', 'hhrhid': 'id', 'tifditp': 'FY disposable income (pos)', 'tifditn': 'FY disposable income (neg)', 'opnum':'extra properties'}, inplace=True)
```

```python
clean['owns property'].value_counts()
```

```python
clean[clean['wave'] == 18.0].plot.scatter(x='education', y ='extra properties')
```

```python
# Group data by 'education' and 'owns property', and count occurrences
education_property_counts = clean[clean['wave'] == 18.0].groupby(['education', 'owns property']).size().unstack()

# Plot the bar graph
education_property_counts.plot(kind='bar', stacked=True)
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Distribution of Education Level by Property Ownership in 2019')
plt.xticks(rotation=45)
plt.legend(title='Owns Property', labels=['No', 'Yes'])
plt.show()
```

# How much are people paying for their mortgages or rent as a proportion of their income overtime?

```python
clean['mortgage monthly'] = 0
```

```python
mortgage_mapping = {
    '[0] No mortgage/all repaid': 0,
    '[-1] Not asked': np.nan,
    '[-3] Dont know': np.nan,
    '[-4] Refused/Not stated': np.nan
    
}

clean['mortgage monthly'] = hilda['hsmg'].replace(mortgage_mapping)
```

```python
clean['mortgage monthly'] = clean['mortgage monthly'].apply(pd.to_numeric, errors='coerce')
```

```python
# Plot histogram of 'mortgage monthly' variable
plt.hist(clean[clean['wave'] == 21.0]['mortgage monthly'], bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Mortgage Monthly')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Mortgage Monthly')

# Show plot
plt.show()
```

```python
income_to_mortgage = []
for i in clean.index:
    if clean.loc[i, 'wave'] == 21.0:
        if clean.loc[i, 'mortgage monthly'] != 0 and not np.isnan(clean.loc[i, 'mortgage monthly']) and clean.loc[i, 'weekly work income'] != 0:
            mortgage = clean.loc[i, 'mortgage monthly']
            income = clean.loc[i, 'weekly work income']*4
            ratio = mortgage/income
            income_to_mortgage.append(ratio)
```

```python
income_to_mortgage_series = pd.Series(income_to_mortgage)
```

```python
# Plot KDE plot of income_to_mortgage ratio
sns.kdeplot(income_to_mortgage, color='skyblue')

# Add labels and title
plt.xlabel('Income to Mortgage Ratio')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Income to Mortgage Ratio 2022')

# Show plot
plt.show()
```

```python
income_to_mortgage_series.describe()
```

```python
# Dictionary to store series for each wave
income_to_mortgage_series = {}
average = []

# Iterate through waves up to 21
for wave in range(1, 22):
    # Initialize empty list for income_to_mortgage ratio
    income_to_mortgage = []
    sum = 0.0
    count = 0
    
    # Filter data for the current wave
    wave_data = clean[clean['wave'] == wave]
    
    # Calculate income_to_mortgage ratio for each row in the current wave
    for i in wave_data.index:
        if wave_data.loc[i, 'mortgage monthly'] != 0 and not np.isnan(wave_data.loc[i, 'mortgage monthly']) and wave_data.loc[i, 'weekly work income'] != 0:
            mortgage = wave_data.loc[i, 'mortgage monthly']
            income = wave_data.loc[i, 'weekly work income'] * 4
            ratio = mortgage / income
            income_to_mortgage.append(ratio)
            sum += ratio
            count += 1

    average.append(sum/count)
    
    # Convert list to pandas Series and store it in the dictionary
    income_to_mortgage_series[f'Wave {wave}'] = pd.Series(income_to_mortgage)
```

```python
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))

# Iterate through income_to_mortgage_series dictionary and plot KDE for each wave
for wave, series in income_to_mortgage_series.items():
    sns.kdeplot(series, ax=ax, label=wave)

# Add labels and title
ax.set_xlabel('Income to Mortgage Ratio')
ax.set_ylabel('Density')
ax.set_title('Kernel Density Estimation of Income to Mortgage Ratio by Wave')

# Set x-axis limits
ax.set_xlim(-3, 3)

# Add legend
ax.legend(title='Wave')

# Show plot
plt.show()
```

```python
average_series = pd.Series(average)

# Modify the index to change the x-axis data points
# For example, if you want to shift the data points by 1
new_index = average_series.index + 1  # Increment the index by 1

# Assign the new index to the Series
average_series.index = new_index

# Plot the Series
average_series.plot(marker='o', linestyle='-')

# Add labels and title
plt.xlabel('New X-axis Data Points')
plt.ylabel('Average Value')
plt.title('Averages Across Waves')

# Show plot
plt.show()
```
